## Querying the Datacube database <a id='Querying'></a>

Define the `DatasetTypes`, min/max `acquistion date period`, and the `Cell` of interest from which to base the query on.

```python
dataset_types = [DatasetType.ARG25, DatasetType.WATER]
```

```python
min_date = date(2000, 01, 01)
max_date = date(2000, 12, 31)
```

```python
satellites = [Satellite(i) for i in ['LS5', 'LS7']]
```

```python
x_cell = [149]
y_cell = [-36]
```

Execute the query and return a `list` of `datasets` ordered by time (oldest to newest).

```python
tiles = list_tiles_as_list(x=x_cell, y=y_cell, acq_min=min_date,
                           acq_max=max_date,
                           satellites=satellites,
                           dataset_types=dataset_types)
```

```python
print "Number of time periods: {}".format(len(tiles))
```

```python
print "Returned DatasetTypes:"
print
print tiles[0].datasets
```

Retrieve the `NBAR` and `WATER` datasets, and their associated band name definitions.

```python
nbar_ds = tiles[2].datasets[DatasetType.ARG25]
water_ds = tiles[2].datasets[DatasetType.WATER]
bands = nbar_ds.bands
water_band = water_ds.bands
metadata = get_dataset_metadata(nbar_ds)
```

Retrieving data

```python
data = get_dataset_data(nbar_ds)
water = get_dataset_data(water_ds)[water_band.WATER]
dtype = data[bands.BLUE].dtype
```

```python
samples, lines = metadata.shape
print "Array dimensions:\nx: {x} & y: {y}".format(x=samples, y=lines)
```

## Image display <a id='Display'></a>

```python
rgb = numpy.zeros((lines, samples, 3), dtype=dtype)
```

```python
rgb[:, :, 0] = data[bands.NEAR_INFRARED]
rgb[:, :, 1] = data[bands.RED]
rgb[:, :, 2] = data[bands.GREEN]
imshow(rgb)
```

It doesn't look like much, so lets scale the data using a histogram equalisation process.

```python
scaled_rgb = numpy.zeros(rgb.shape, dtype='uint8')
for i in range(3):
    scaled_rgb[:, :, i] = hist_equal(rgb[:, :, i], binsize=1)
```

```python
imshow(scaled_rgb)
```

The `WATER` `dataset` is a classification with classes defined at the values `0`, `1`, `2`, `4`, `8`, `16`, `32`, `64` & `128`.

```python
imshow(water)
```

## Basic Water classification (threshold) <a id='Threshold'></a>

Compute a histogram of the `NEAR_INFRARED` band.

```python
h = histogram(data[bands.NEAR_INFRARED], minv=0)
hist = h['histogram']
```

```python
plot(hist)
```

There appears to be some sort of bimodality, so we'll take a threshold of the `NEAR_INFRARED` band at the value 1200.

```python
plot(hist)
plot([1200, 1200], [0, hist.max()], 'r-')
```

```python
thresh = (rgb[:, :, 0] > 0) & (rgb[:, :, 0] < 1200)
```

Display the threshold

```python
imshow(thresh, cmap=cm.gray)
```

## Subsetting <a id='Subsetting'></a>

We'll create a `Region of Interest` so we can take a closer look at the regions defined by the threshold.

```python
roi = ((0, 1500), (0, 2000))
ys, ye = roi[0]
xs, xe = roi[1]
```

Subset the `water`, `thresheld` and the `rgb` arrays

```python
subs_water = water[ys:ye, xs:xe]
subs_thresh = thresh[ys:ye, xs:xe]
rgb_subs = rgb[ys:ye, xs:xe, :]
```

Initialise an empty array that we can use to contain a scaled `RGB` of the subsetted `RGB` array.  This time we'll enhance the image via a percentile.

```python
scaled_rgb = numpy.zeros(rgb_subs.shape, dtype='uint8')
```

```python
for i in range(3):
    scaled_rgb[:, :, i] = hist_equal(rgb_subs[:, :, i], binsize=1, percent=3)
```

```python
imshow(scaled_rgb)
```

Display the classified image derived from the `WOfS` algorithm.

```python
imshow(subs_water, cmap=cm.gray)
```

Display our basic threshold classfier.  Looks pretty close.

```python
imshow(subs_thresh, cmap=cm.gray)
```

## Segmentation analysis <a id='Segment-Analysis'></a>

### Defining segments <a id='Define-segments'></a>

What we'll do here is analyse each of the water segments returned via the `WOfS` algorithm and our basic threshold.

The `label_region` funtion is a pixel connectivity function that labels pixels based on neighbourhood connectivity.  The default behaviour is to connect pixels via the 4 immediate neighbours, and assign a label/segment ID.

```python
water_lab = label_region(subs_water == 128) # Water is classified as the value 128
```

```python
thresh_lab = label_region(subs_thresh)
```

```python
imshow(water_lab)
```

```python
imshow(thresh_lab)
```

### SegmentVisitor <a id='SegmentVisitor'></a>

Using the `SegmentVisitor`, we can quickly analyse each of the segments.

```python
water_seg = SegmentVisitor(water_lab)
thresh_seg = SegmentVisitor(thresh_lab)
print "Number of water segments: {}".format(water_seg.n_segments)
print "Number of threshold segments: {}".format(thresh_seg.n_segments)
```

There are probably a lot of little segments not worth investigating here.

```python
print "Number of water segments < 5 pixels in size: {}".format(thresh_seg.histogram[water_seg.histogram < 5].shape[0])
print "Number of threhold segments < 5 pixels in size: {}".format(thresh_seg.histogram[thresh_seg.histogram < 5].shape[0])
```

What we can do is sieve the segments to remove any segments that contain less than 5 pixels.

```python
thresh_seg.sieve_segments(5, inplace=True)
water_seg.sieve_segments(5, inplace=True)
```

```python
print "Number of water segments < 5 pixels in size: {}".format(thresh_seg.histogram[thresh_seg.histogram < 5].shape[0])
print "Number of threshold segments < 5 pixels in size: {}".format(thresh_seg.histogram[thresh_seg.histogram < 5].shape[0])
```

How many segments are we left with?

```python
print "Number of water segments: {}".format(water_seg.n_segments)
print "Number of threshold segments: {}".format(thresh_seg.n_segments)
```

What are our segment ID's?

```python
water_seg.segment_ids
```

Now lets do some basic analysis on each of the segments.

For the `water` `dataset`, we'll only calculate the `mean` value for each segment.

```python
water_nir_mean = water_seg.segment_mean(rgb_subs[:, :, 0])
water_red_mean = water_seg.segment_mean(rgb_subs[:, :, 1])
water_green_mean = water_seg.segment_mean(rgb_subs[:, :, 2])
```

The current return structure is a dictionary with the `keys` as the segment ID's. A later release will include an option to return a `pandas.DataFrame`.  But for now, let's construct one of our own.

```python
water_nir_mean
```

Create a `pandas.DataFrame` and populate it with our `mean` value for each segment.

```python
water_df = pandas.DataFrame(index=water_nir_mean.keys())
water_df['Mean_NIR'] = pandas.DataFrame.from_dict(water_nir_mean, orient='index')
water_df['Mean_RED'] = pandas.DataFrame.from_dict(water_red_mean, orient='index')
water_df['Mean_GREEN'] = pandas.DataFrame.from_dict(water_green_mean, orient='index')
```

```python
water_df.head(10)
```

#### Analysis of segments defined by threshold <a id='Analyse_segments'></a>

Now let's do the same for our threshold example.

Here we'll calculate the `mean`, `standard deviation`, `min`, `max` and `area` for each segment.

```python
# Mean
thresh_nir_mean = thresh_seg.segment_mean(rgb_subs[:, :, 0])
thresh_red_mean = thresh_seg.segment_mean(rgb_subs[:, :, 1])
thresh_green_mean = thresh_seg.segment_mean(rgb_subs[:, :, 2])

# StdDev
thresh_nir_stddev = thresh_seg.segment_stddev(rgb_subs[:, :, 0])
thresh_red_stddev = thresh_seg.segment_stddev(rgb_subs[:, :, 1])
thresh_green_stddev = thresh_seg.segment_stddev(rgb_subs[:, :, 2])

# Min
thresh_nir_min = thresh_seg.segment_min(rgb_subs[:, :, 0])
thresh_red_min = thresh_seg.segment_min(rgb_subs[:, :, 1])
thresh_green_min = thresh_seg.segment_min(rgb_subs[:, :, 2])

# Max
thresh_nir_max = thresh_seg.segment_max(rgb_subs[:, :, 0])
thresh_red_max = thresh_seg.segment_max(rgb_subs[:, :, 1])
thresh_green_max = thresh_seg.segment_max(rgb_subs[:, :, 2])

# Area
thresh_area = thresh_seg.segment_area(scale_factor=629.135776) # Avg pixel area in square metres for this cell
```

Create a `pandas.DataFrame` and populate it with our results.

```python
df = pandas.DataFrame({'Seg_ID': thresh_seg.segment_ids}, index=range(1, thresh_seg.n_segments + 1))

# Mean
df['Mean_NIR'] = pandas.DataFrame.from_dict(thresh_nir_mean, orient='index')
df['Mean_RED'] = pandas.DataFrame.from_dict(thresh_red_mean, orient='index')
df['Mean_GREEN'] = pandas.DataFrame.from_dict(thresh_green_mean, orient='index')

# StdDev
df['StdDev_NIR'] = pandas.DataFrame.from_dict(thresh_nir_stddev, orient='index')
df['StdDev_RED'] = pandas.DataFrame.from_dict(thresh_red_stddev, orient='index')
df['StdDev_GREEN'] = pandas.DataFrame.from_dict(thresh_green_stddev, orient='index')

# Min
df['Min_NIR'] = pandas.DataFrame.from_dict(thresh_nir_min, orient='index')
df['Min_RED'] = pandas.DataFrame.from_dict(thresh_red_min, orient='index')
df['Min_GREEN'] = pandas.DataFrame.from_dict(thresh_green_min, orient='index')

# Max
df['Max_NIR'] = pandas.DataFrame.from_dict(thresh_nir_max, orient='index')
df['Max_RED'] = pandas.DataFrame.from_dict(thresh_red_max, orient='index')
df['Max_GREEN'] = pandas.DataFrame.from_dict(thresh_green_max, orient='index')

# Area
df['Area_m2'] = pandas.DataFrame.from_dict(thresh_area, orient='index')

# Reset the index to start from zero
df.reset_index(drop=True, inplace=True)
```

```python
df.head(10)
```

### Querying the analysed segments <a id='Pandas-queryies'></a>

Now that we have a the data contained within a `pandas.DataFrame`, we can so some queries and return the segments that satisfy our query.

For this example, we'll select only those segments that have a `standard deviation` greater than 220 for the `NEAR_INFRARED` band, and an `area` that is greater than 40000m<sup>2.

```python
mask = (df['StdDev_NIR'] > 220) & (df['Area_m2'] > 40000)
result = df[mask]
```

```python
result
```

Retrieve the bounding box for the segment. This will be used to index the `RGB` subset image, thereby producing another subset.  But this time, the subset should be centred around the segment 206.

```python
bbox = thresh_seg.segment_bounding_box([206])
```

```python
print "Bounding box: ", bbox
```

Get the starting and ending co-ordinates for both the `x` and `y` dimensions, and index the image.

```python
ys, ye = bbox[206][0]
xs, xe = bbox[206][1]
segment_img = rgb_subs[ys:ye, xs:xe, :]
```

Scale the image and display

```python
scl_seg_img = numpy.zeros(segment_img.shape, dtype='uint8')
for i in range(3):
    scl_seg_img[:, :, i] = hist_equal(segment_img[:, :, i], binsize=1, percent=1)
```
