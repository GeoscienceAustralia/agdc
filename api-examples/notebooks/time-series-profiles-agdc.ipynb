# An example of retrieving timeseries for cropping and non-cropping landcovers

```python
from datetime import date
import json
import matplotlib.cm as cm
import numpy
from datacube.api.model import DatasetType, Satellite
from datacube.api.query import list_tiles_as_list
from datacube.api.utils import get_dataset_data, get_dataset_metadata
from datacube.api.utils import get_dataset_data_stack, latlon_to_xy, DEFAULT_MASK_PQA
import matplotlib.cm as cm
import pandas
```

Define some basic properties of the acquisitions that we want to query.

```python
dataset_types = [DatasetType.ARG25, DatasetType.WATER, DatasetType.FC25, DatasetType.PQ25]
min_date = date(1985, 01, 01)
max_date = date(2015, 12, 31)
satellites = [Satellite(i) for i in ['LS5', 'LS7', 'LS8']]
x_cell = [146]
y_cell = [-35]
```

```python
tiles = list_tiles_as_list(x=x_cell, y=y_cell, acq_min=min_date,
                           acq_max=max_date,
                           satellites=satellites,
                           dataset_types=dataset_types)
```

```python
print "Number of tiles/time slices: {}".format(len(tiles))
```

```python
nbar_ds = tiles[0].datasets[DatasetType.ARG25]
bands = nbar_ds.bands
metadata = get_dataset_metadata(nbar_ds)
geotransform = metadata.transform
```

We have two lat/lon co-ordinates of interest, so convert them to image/array co-ordinates

```python
imgx1, imgy1 = latlon_to_xy(-34.1059362, 146.1436772, geotransform)
imgx2, imgy2 = latlon_to_xy(-34.0172117, 146.1882717, geotransform)
print imgx1, imgy1
print imgx2, imgy2
```

Extract the timeseries for the `NEAR_INFRARED` & `RED` bands for each of the lat/long co-ordinates of interest.

What is returned is a `list` of 2D arrays, so we'll convert to a `NumPy` and flatten to a 1D array.

We'll also get the tool to extract the `PQ` information and mask out pixels flagged by the `PQ` band.

```python
nir_ts1 = numpy.array(get_dataset_data_stack(tiles, DatasetType.ARG25, bands.NEAR_INFRARED.name, x=imgx1, y=imgy1, x_size=1, y_size=1,
                                             mask_pqa_apply=True, mask_pqa_mask=DEFAULT_MASK_PQA), dtype='float').flatten()
```

```python
red_ts1 = numpy.array(get_dataset_data_stack(tiles, DatasetType.ARG25, bands.RED.name, x=imgx1, y=imgy1, x_size=1, y_size=1,
                                             mask_pqa_apply=True, mask_pqa_mask=DEFAULT_MASK_PQA), dtype='float').flatten()
```

```python
nir_ts2 = numpy.array(get_dataset_data_stack(tiles, DatasetType.ARG25, bands.NEAR_INFRARED.name, x=imgx2, y=imgy2, x_size=1, y_size=1,
                                             mask_pqa_apply=True, mask_pqa_mask=DEFAULT_MASK_PQA), dtype='float').flatten()
```

```python
red_ts2 = numpy.array(get_dataset_data_stack(tiles, DatasetType.ARG25, bands.RED.name, x=imgx2, y=imgy2, x_size=1, y_size=1,
                                             mask_pqa_apply=True, mask_pqa_mask=DEFAULT_MASK_PQA), dtype='float').flatten()
```

The `no-data` value for this `DatasetType` is `-999`, and any pixels masked out by the `PQ` band will be set to `-999` as well. So this step is to turn any pixels containing `-999` into a `NaN`.

```python
nir_ts1[nir_ts1 == -999] = numpy.nan
red_ts1[red_ts1 == -999] = numpy.nan
nir_ts2[nir_ts2 == -999] = numpy.nan
red_ts2[red_ts2 == -999] = numpy.nan
```

We'll also extract the acquisition `datetimes`. This will be useful for when we load the data into `pandas`.

```python
timestamps = []
for tile in tiles:
    timestamps.append(tile.start_datetime)
```

```python
data = {'NIR_Dryland_Cropping': nir_ts1,
        'RED_Dryland_Cropping': red_ts1,
        'NIR_Open_Bushland': nir_ts2,
        'RED_Open_Bushland': red_ts2}
```

```python
df = pandas.DataFrame(data, index=timestamps)
```

```python
df.head(10)
```

In this example we'll comute an `NDVI` using the extracted time series for both of the landcover types.

```python
df['NDVI_Dryland_Cropping'] = df.eval("(NIR_Dryland_Cropping - RED_Dryland_Cropping) / (NIR_Dryland_Cropping + RED_Dryland_Cropping)")
```

```python
df['NDVI_Open_Bushland'] = df.eval("(NIR_Open_Bushland - RED_Open_Bushland) / (NIR_Open_Bushland + RED_Open_Bushland)")
```

```python
df.head(5)
```

Here we'll subset the `DataFrame` and create a new one that contains only the computed `NDVI's`.

```python
ndvi_df = df.loc[:, ['NDVI_Dryland_Cropping', 'NDVI_Open_Bushland']]
```

We can plot the entire timeseries

```python
ndvi_df.plot(title='NDVI Timeseries')
```

Or we can plot a subset of the timeseries

```python
ndvi_df['2003':'2007'].plot(title='2003-2007 NDVI Timeseries')
```

Another fantastic feature of pandas is its ability to handle timeseries data and resample it one sees fit.

```python
ndvi_df.resample('Q').plot(title='NDVI Timeseries Resampled Quarterly')
```
