# An example to retrieve the data that falls under a line contained within a vector file

```python
from datetime import date
import json
import matplotlib.cm as cm
import numpy
import ogr
import pandas
import sys
sys.path.append('/home/547/jps547/git_repositories/sixy6e/agdc/api/source/main/python/')
from datacube.api.model import DatasetType, Satellite, BANDS
from datacube.api.query import list_tiles_as_list, list_tiles_wkt
from datacube.api.utils import get_dataset_data, get_dataset_metadata, arbitrary_profile
from eotools.coordinates import convert_coordinates
from idl_functions import bytscl
```

Using the `ogr` library, load the `shapefile` that contains our line-vector of interest.

```python
vec_fname = '/g/data2/v10/testing_ground/jps547/api-transects/LBG-transect_redo.shp'
vec_ds = ogr.Open(vec_fname)
layer = vec_ds.GetLayer(0)
feat = layer.GetFeature(0)
ref = feat.GetGeometryRef()
```

Define some basic properties of the acquisitions that we want to query.

```python
dataset_types = [DatasetType.ARG25]
min_date = date(1995, 01, 01)
max_date = date(1995, 12, 31)
years = range(min_date.year, max_date.year+1)
satellites = [Satellite('LS5')]
```

Using the vector file we loaded previously, we can directly query the `agdc` and return a list of the aquisitions tiles that meet our criteria.

```python
tiles = list_tiles_wkt(ref.ExportToWkt(), satellites=satellites, years=years, datasets=dataset_types)
```

For this example, we'll investigate the second tile from the list, and retrieve some basic properties such as the `DatasetType`, the available bands and the UL corner co-ordinate tie-point.

```python
dataset = tiles[1].datasets[DatasetType.ARG25]
bands = dataset.bands
metadata = get_dataset_metadata(dataset)
geotransform = metadata.transform
```

Using the information from the loaded vector file, retrieve the extent/bounding box. We'll use this to create an image subset from the `ARG25` dataset.

```python
extent = layer.GetExtent()
print "Extent: {}".format(extent)
```

The `extent` information will be used to create an `ROI` that can be used as an index when reading the data from disk.

```python
UL_map_xy = (extent[0], extent[3])
LR_map_xy = (extent[1], extent[2])
UL_imgx, UL_imgy = convert_coordinates(geotransform, UL_map_xy, to_map=False)
LR_imgx, LR_imgy = convert_coordinates(geotransform, LR_map_xy, to_map=False)
roi_tile = ((UL_imgy, LR_imgy), (UL_imgx, LR_imgx))
print "ROI subset: {}".format(roi_tile)
```

Extract the starting and ending co-ordinates, and the block size.

```python
ys, ye = roi_tile[0]
xs, xe = roi_tile[1]
ysize = ye - ys
xsize = xe - xs
```

To display an RGB image, we'll extract the `NEAR_INFRARED`, `RED`, & `GREEN` bands.

```python
rgb_bands = [bands.NEAR_INFRARED, bands.RED, bands.GREEN]
```

Once the data has been read from disk, a `dictionary` is returned as the container, and the values of the returned dictionary (`NumPy` arrays in this case), can be accessed via the band names.

```python
data = get_dataset_data(dataset, rgb_bands, x=xs, y=ys, x_size=xsize, y_size=ysize)
```

```python
roi_subset = numpy.zeros((3, roi_tile[0][1] - roi_tile[0][0], roi_tile[1][1] - roi_tile[1][0]), dtype='uint8')
roi_subset[0] = bytscl(data[bands.NEAR_INFRARED], minv=1)
roi_subset[1] = bytscl(data[bands.RED], minv=1)
roi_subset[2] = bytscl(data[bands.GREEN], minv=1)
```

Now to access the vertices of the line/vector contained within the `shapefile`.

Convert the geometry in JSON format. This format makes it easier to handle co-ordinate conversions.

```python
line = json.loads(ref.ExportToJson())
map_coords = line['coordinates']
```

Convert the real world co-ordinates into image co-ordinates.

```python
img_coords = []
for coord in map_coords:
    img_coords.append(convert_coordinates(geotransform, coord, to_map=False))
```

Have a quick look at how complex the transect is. The more vertices the more complex and potentially longer profile compute times. More complex transects should avoid setting the `cubic=True` argument in the `arbitrary_profile` function. Default is `False` in which case `nearest neighbour` interpolation will be used.

```python
print "No. of vertices: {}".format(len(img_coords))
```

Compute the image offsets introduced during the subset creation. This'll allow the display of the vector over the top of the subsetted image.

```python
img_coords_offset = [(x - roi_tile[1][0], y - roi_tile[0][0]) for (x, y) in img_coords]
x_off = [x for (x, y) in img_coords_offset]
y_off = [y for (x, y) in img_coords_offset]
```

Retreive the transect. If no `band` is provided, the default behaviour is to read the first band in the list returned by `dataset.bands`.

```python
blue_transect, idx, xy_start_end = arbitrary_profile(dataset, map_coords, from_map=True)
```

In this example, we'll create a transect from the `NEAR_INFRARED` band.

```python
nir_transect, idx, xy_start_end = arbitrary_profile(dataset, map_coords, band=bands.NEAR_INFRARED, from_map=True)
```

Display the image subset and overlay the transect.

```python
imshow(roi_subset.transpose(1, 2, 0), cmap=cm.Greys_r)
axis('image')
plot(x_off, y_off, 'g-')
```

Plot both the `BLUE` & the `NEAR_INFRARED` transects/profiles.

```python
plot(blue_transect, 'b-')
plot(nir_transect, 'r-')
```
